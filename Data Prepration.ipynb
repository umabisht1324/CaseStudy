```python
import pandas as pd
```

```python
filePath="Command Centre Dataset.xlsx"
xls=pd.ExcelFile(filePath)
```

```python
xls.sheet_names
```

```python
sales_data=pd.read_excel(xls,sheet_name="Sales_Data")
marketing_data=pd.read_excel(xls,sheet_name="Marketing_Data")
customer_feedback_data=pd.read_excel(xls,sheet_name="Customer_Feedback_Data")
competitor_data=pd.read_excel(xls,sheet_name="Competitor_Data")
```

```python
marketing_data.head()
```

# DATA  CLEANING 

## Cleaning Sales Data
<!-- ### 1. Removing Duplicates
### 2. Removing Rows with NULL values
### 3. Standardize Date Format -->

```python
sales_data.drop_duplicates(inplace=True)  
sales_data.dropna(inplace=True)  
sales_data['Date'] = pd.to_datetime(sales_data['Date']) 
```

```python
sales_data.head()
```

## Cleaning Marketing Data
<!-- ### 1. Removing Duplicates
### 2. Removing Rows with NULL values
### 3. Standardize Date Format -->

```python
marketing_data.drop_duplicates(inplace=True)
marketing_data.dropna(inplace=True)
marketing_data['Date']=pd.to_datetime(marketing_data['Date'])
```

```python
marketing_data.head()
```


## Cleaning Customer Feedback Data

<!-- ### 1. Removing Duplicates
### 2. Removing Rows with NULL values
### 3. Standardize Date Format -->

```python
customer_feedback_data.drop_duplicates(inplace=True)
customer_feedback_data.dropna(inplace=True)
customer_feedback_data['Date']=pd.to_datetime(customer_feedback_data['Date'])
```

```python
customer_feedback_data.head()
```

## Cleaning Competitor Data


```python
competitor_data.drop_duplicates(inplace=True)
competitor_data.dropna(inplace=True)
```

```python
competitor_data.head()
```

# STANDARDIZE  COLUMN  NAMES

```python
sales_data.rename(columns=lambda x:x.lower(),inplace=True) 
marketing_data.rename(columns=lambda x:x.lower(),inplace=True)
customer_feedback_data.rename(columns=lambda x:x.lower(),inplace=True)
competitor_data.rename(columns=lambda x:x.lower(),inplace=True)
```

```python
competitor_data.head()
```

# MERGING  DATASET

```python
merged_data=sales_data.merge(marketing_data, on=['country','zone','date'],how='inner')
merged_data=merged_data.merge(customer_feedback_data, on=['country','zone','product_category','date'],how='inner')
merged_data=merged_data.merge(competitor_data, on=['country','zone'], how='left')



```

```python
merged_data.head()
```

# SAVING  THE  CLEANED  DATASET

```python
merged_data.to_csv("Cleaned_Command_Centre_Dataset.csv", index=False)
```
